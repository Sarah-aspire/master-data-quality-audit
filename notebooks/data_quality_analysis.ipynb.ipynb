```python
import pandas as pd

# Load Excel dataset

df = pd.read_excel(r"C:\Users\sarah25\Downloads\product_service_master_data.xlsx")

# Preview dataset

df.head()
```

```python
# Dataset overview

print("Rows and Columns:")
print(df.shape)

print("\nColumn Information:")
print(df.info())
```

```python
# Missing values analysis

missing_values = df.isnull().sum()

print("Missing Values:")
print(missing_values)
```

```python
# Duplicate service codes

duplicates = df[df.duplicated(subset=["Service_Code"], keep=False)]

print("Duplicate Service Codes:")
duplicates
```

```python
# Invalid status values

valid_status = ["Active", "Inactive", "Pending"]

invalid_status = df[~df["Status"].isin(valid_status)]

print("Invalid Status Values:")
invalid_status
```

```python
# Incorrect cost center format

invalid_costcenter = df[
~df["Cost_Center"].astype(str).str.startswith("CC-")
]

print("Incorrect Cost Center Format:")
invalid_costcenter
```

```python
# Data Quality KPI Summary

summary = {
"Total Records": len(df),
"Duplicate Service Codes": duplicates["Service_Code"].nunique(),
"Missing Cost Centers": df["Cost_Center"].isnull().sum(),
"Missing Responsible Teams": df["Responsible_Team"].isnull().sum(),
"Invalid Status Values": len(invalid_status)
}

summary_df = pd.DataFrame(
summary.items(),
columns=["KPI", "Value"]
)

summary_df
```
